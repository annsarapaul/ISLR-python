```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wage = pd.read_csv('Data/Wage.csv', index_col = 0)
wage.head(5)
```

```python
wage.info()
```

```python
mean_wage_by_age = wage[['age','wage']].groupby('age').mean()
sns.scatterplot(data=wage, x="age", y="wage")
plt.plot(mean_wage_by_age.index, mean_wage_by_age['wage'], color='b')
plt.show()
```

```python
sns.relplot(x="age", y="wage",
            dashes=False, markers=True, kind="line", data=wage);
```

```python
mean_wage_by_year = wage[['year','wage']].groupby('year').mean()
sns.scatterplot(data=wage, x="year", y="wage")
plt.plot(mean_wage_by_year.index, mean_wage_by_year['wage'], color='b')
plt.show()
```

```python
ax = sns.boxplot(x="education", y="wage", data=wage)
```

```python
advertising = pd.read_csv('Data/Advertising.csv',index_col=0)
advertising.head()
```

```python
channels = ['TV', 'Radio','Newspaper']
fig, axs = plt.subplots(1,len(channels))
fig.set_size_inches(18.5, 10.5)
plt.figure(1)
i = 0

for channel in channels:
    #axs[i] = sns.scatterplot(data=advertising, x=channel, y="Sales")
    sns.scatterplot(data=advertising, x=channel, y="Sales",ax=axs[i]).set_title(channel)
    m, b = np.polyfit(advertising[channel], advertising['Sales'], 1)
    axs[i].plot(advertising[channel], m*advertising[channel] + b,color='b')
    i = i +1
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(advertising['Radio'], advertising['TV'], advertising['Sales'], cmap=plt.cm.viridis, linewidth=0.2)
#plt.plot(advertising[channel], m*advertising[channel] + b,color='r')

plt.show()
```
